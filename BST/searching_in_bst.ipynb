```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def insert(node, val):
    if node is None:
        return Node(val)
    elif(val < node.value):
        node.left = insert(node.left, val)
    else:
        node.right = insert(node.right,val)
    
    return node
```

```python
root = Node(1)
```

```python
arr = [1,2,3,4,5,6,7,8,1,23,4,53,8,0]
```

```python
for i in arr[1:]:
    root = insert(root,i)
```

```python
def print_tree(nod):
    if nod is None:
        return
    print(f"{nod.value} ")
    print_tree(nod.left)
    print_tree(nod.right)
```

```python
print_tree(root)
```

```python
def search_tree(nod, target):
    if nod is None:
        return False
        
    if nod.value == target:
        return True
        
    if target < nod.value:
        return search_tree(nod.left, target) 
    else:
        return search_tree(nod.right, target)
```

```python
ans = search_tree(root, 23)
```

```python
print(ans)
```
